```python
df.fillna({"Price": df["Price"].mean(),
           "Discount": df["Discount"].mean(),
           "Profit": df["Profit"].mean()}, inplace=True)
df
```

```python
df.dropna(subset=["Customer_Name"], inplace=True)  
df.drop(df[df["Customer_Name"] == "Unknown"].index, inplace=True)  
df.drop(df[df["Customer_Name"] == ""].index, inplace=True) 
df
```

```python
sns.barplot(
    x=["Total Sales", "Total Profit"], 
    y=[df["Total_Sales"].sum(), df["Profit"].sum()], 
    hue=["Total Sales", "Total Profit"],  
    palette="viridis", 
    legend=False  
)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Discount"], y=df["Profit"], alpha=0.7, color="purple")
plt.title("Discount vs Profit")
plt.xlabel("Discount (%)")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df["Region"], y=df["Total_Sales"])
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Month"] = df["Order_Date"].dt.month

plt.figure(figsize=(10, 5))
sns.lineplot(x="Month", y="Total_Sales", data=df, marker="o", color="green")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_customers = df.groupby("Customer_Name")["Total_Sales"].sum().nlargest(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_customers.index, x=top_customers.values)
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Name")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(y=df["Product"], order=df["Product"].value_counts().index[:10])
plt.title("Top 10 Most Sold Products")
plt.xlabel("Count")
plt.ylabel("Product")
plt.show()
```
